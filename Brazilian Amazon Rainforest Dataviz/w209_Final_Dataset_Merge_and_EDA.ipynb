## Deforestation Area (km²) by Year and State

Deforestation area (km²) by year and state, from 2004 to 2019. The data are public and were extracted from INPE website on December 16th 2019. It was already aggregated, so, no data process was made. Program: PRODES (Programa de Monitoramento da Floresta Amazônica Brasileira por Satélite, or Brazilian Amazon Rainforest Monitoring Program by Satellite). Methodology: maps primary forest loss using satellite imagery, with 20 to 30 meters of spatial resolution and 16-day revisit rate, in a combination that seeks to minimize the problem of cloud cover and ensure interoperability criteria.

```
def_area_2004_2019.head()
```

```
def_area_2004_2019 = def_area_2004_2019.drop(columns=['AMZ LEGAL'])
def_area_2004_2019.head()
```

```
def_area_2004_2019 = def_area_2004_2019.melt(id_vars="Ano/Estados", var_name="state",value_name="deforested_area")
def_area_2004_2019.head()
```

```
def_area_2004_2019['state'].unique()
```

## Amount of Fire Outbreaks in Brazilian Amazon

Amount of fire outbreaks in Brazilian Amazon by state, month and year, from 1999 to 2019. The original data are public and were extracted from INPE website on December 13th 2019, always from the filtering for the reference salellite and aggregated using Postgres SQL so that you could work with lighter files. Program: BDQ (Banco de Dados de Queimadas, or Fires Database). Methodology: detects fire outbreaks through satellite images, updated every 3 hours.

## Mapping table

```
mappping_df = pd.DataFrame({'state': ['ACRE', 'AMAZONAS', 'AMAPA', 'MARANHAO', 'MATO GROSSO', 'PARA', 'RONDONIA', 'RORAIMA', 'TOCANTINS'], 'state_abb': ['AC', 'AM', 'AP', 'MA', 'MT', 'PA', 'RO', 'RR', 'TO']})
```

```
initial_merge = pd.merge(inpe_brazilian_amazon_fires_1999_2019, mappping_df, how='left', on=['state', 'state'])
```

```
initial_merge.head()
```

```
#change column for merge
def_area_2004_2019.rename(columns={'state':'state_abb', 'Ano/Estados':'year'}, inplace=True)
```

```
#need to merge by year then state_abb
second_merge = initial_merge.merge(def_area_2004_2019, how='left', left_on=['year', 'state_abb'], right_on=['year', 'state_abb'])
```

```
second_merge.head()
```

```
final_merge = second_merge.merge(el_nino_la_nina_1999_2019, how='left', left_on=['year'], right_on=['start year'])
```

```
final_merge.head()
```

## Rainforest Degredation by Year

```
df = final_merge[final_merge['deforested_area'].notna()]
```

```
alt.Chart(df).mark_line().encode(
    alt.X('year:N',
    ),
    y='sum(deforested_area):Q',
    color = 'state'
).properties(
    title='Rainforest Degredation by Year'
)
```

```
alt.Chart(df).mark_area().encode(
    alt.X('year:N'),
    alt.Y('sum(deforested_area):Q'),
    color='state',
    #column='state'
).properties(
    width=250,
    height=250
).facet(
    'state',
    columns = 3
)
```
